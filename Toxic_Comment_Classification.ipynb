```python
from tensorflow.keras.layers import TextVectorization
```

```python
x=df.comment_text
y=df[df.columns[2:]].values
```

```python
x
```

```python
y
```

```python
MAX_FEATURES=200000
```

```python
vectorizer = TextVectorization(max_tokens=MAX_FEATURES,
                               output_sequence_length=2000,
                               output_mode='int')
```

```python
x.values
```

```python
vectorizer.adapt(x.values)
```

```python
vectorizer.get_vocabulary()
```

```python
len(vectorizer.get_vocabulary())
```

```python
vectorizer('Hello world, life is great')
```

```python
vectorized_text=vectorizer(x.values)
```

```python
vectorized_text
```

```python
dataset=tf.data.Dataset.from_tensor_slices((vectorized_text,y))
dataset=dataset.cache()
dataset=dataset.shuffle(160000)
dataset=dataset.batch(16)
dataset=dataset.prefetch(8)
```

```python
dataset
```

```python
batch_x, batch_y = dataset.as_numpy_iterator().next()
```

```python
batch_x.shape
```

```python
len(dataset)*16
```

```python
batch_y.shape
```

```python
train = dataset.take(int(len(dataset)*0.7))
val = dataset.skip(int(len(dataset)*0.7)).take(int(len(dataset)*0.2))
test = dataset.skip(int(len(dataset)*0.9)).take(int(len(dataset)*0.1))
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional, Dense, Embedding
```

```python
model=Sequential()
model.add(Embedding(MAX_FEATURES+1,32))
model.add(Bidirectional(LSTM(32,activation='tanh')))
model.add(Dense(128,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(6,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='BinaryCrossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history = model.fit(train,epochs=20,validation_data=val)
```

```python
from tensorflow.keras.metrics import Precision, Recall
```

```python
pre=Precision()
re=Recall()
```

```python
for batch in test.as_numpy_iterator():
    batch_x, batch_y = batch
    yhat = model.predict(batch_x)

    batch_y = batch_y.flatten()
    yhat = yhat.flatten()

    pre.update_state(batch_y,yhat)
    re.update_state(batch_y,yhat)
```

```python
precision = pre.result().numpy()
```

```python
recall = re.result().numpy()
```

```python
f1score = (2*precision*recall)/(precision+recall)
print(f"precision: {precision}\nrecall: {recall}\nf1score: {f1score}")
```
